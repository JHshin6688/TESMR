```python
#print(int(''))
def load_negative_file(filename):
    negativeList = []
    with open(filename, 'r') as f:
        line = f.readline()
        count = 1
        while line is not None and line != '':
            arr = line.split('\t')
            #print(len(arr), count)
            negatives = []
            for x in arr[1:]:
                negatives.append(int(x))
            negativeList.append(negatives)
            line = f.readline()
            count += 1
    return negativeList

filename = '../../../data/jeehoshin/healthrec_dataset/data.valid.negative'
a = load_negative_file(filename)
#print(a)
```
